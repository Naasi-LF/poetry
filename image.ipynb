```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 400)

y = np.tanh(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.title('tanh Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-10, 10, 400)

y = np.softmax(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.title('tanh Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)
plt.show()
```

```python
def softmax(x):
    if len(x.shape) > 1:
        tmp = np.max(x, axis=1)
        x -= tmp.reshape((x.shape[0], 1))
        x = np.exp(x)
        tmp = np.sum(x, axis=1)
        x /= tmp.reshape((x.shape[0], 1))
    else:
        tmp = np.max(x)
        x -= tmp
        x = np.exp(x)
        tmp = np.sum(x)
        x /= tmp
    return x

x = np.random.rand(10)

softmax_result = softmax(x)
softmax_result

indices = np.arange(len(softmax_result))

plt.figure(figsize=(10, 5))
plt.bar(indices, softmax_result, align='center', alpha=0.5)
plt.xticks(indices)
plt.ylabel('Probability')
plt.title('Softmax Function')

for i in range(len(softmax_result)):
    plt.text(i, softmax_result[i], f'{softmax_result[i]:.2f}', ha='center', va='bottom')
plt.show()
```
